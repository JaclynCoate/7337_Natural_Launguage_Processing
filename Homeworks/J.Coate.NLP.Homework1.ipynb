##### Unit 1 HW: Question 2
Follow the instructions in chapter 1 of Bird-Klein for implementing a “lexical diversity” scoring routine

- Creating a function to repeat the lexical diversity code and percentage word code

```python
#Specific a parameter labeled: text, placeholder for the actual text to use
def lexical_diversity(text):
    return len(text) / len(set(text))
```

```python
def percentage(count, total):
    return 100*count / total
```

```python
#Leveraging our new lexical_diversity function
lexical_diversity(text3)
```

```python
#Leveraging our new lexical_diversity function
lexical_diversity(text5)
```

```python
#Leveraging our new percentage function
percentage(4,5)
```

```python
#Leveraging our new percentage function
percentage(text4.count('a'), len(text4))
```
